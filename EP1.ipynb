```python
import os
import random
import matplotlib.pyplot as plt
from skimage import io
import pandas as pd
import numpy as np

SEED = 323
```

```python
# Usar as imagens armazenadas no Google Drive
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/MAC0417')
```

```python
# Criar dataset dos metadados
data = pd.read_csv('metadados.csv')
data = data.sort_values(by=['Fundo','Local','Iluminacao','Classe']) # Organizar dataset
data = data.reset_index(drop=True)
data
```

```python
# Criar banco de imagens
dataset_f = 'BancoImagens'
imgs_f = [f"{dataset_f}/{img_f}" for img_f in data["Nome do arquivo"]]

#Ler banco de imagens
img_data = io.imread_collection(imgs_f)
img_data
```

```python
# Exemplos de imagens do dataset

imgs_f_random = imgs_f[:]
random.Random(SEED).shuffle(imgs_f_random)

imgs = io.imread_collection(imgs_f_random[:9])

# (fonte: https://stackoverflow.com/questions/19471814/display-multiple-images-in-one-ipython-notebook-cell/27795087)
plt.figure(figsize=(15,15))
columns = 3
for i, img in enumerate(imgs):
    plt.subplot(len(imgs) / columns + 1, columns, i + 1)
    plt.imshow(img)
```
